```python
import os
import pandas as pd
import matplotlib.pyplot as plt

os.makedirs("charts", exist_ok=True)

restaurants = pd.read_csv("fast_food_restaurants.csv")

print(restaurants.columns)

# Count restaurants by state/province
state_counts = (
    restaurants.groupby("province")
    .size()
    .reset_index(name="restaurant_count")
)

print(state_counts.sort_values(
    "restaurant_count",
    ascending=False
).head(20))

# Save table
state_counts.to_csv(
    "charts/question3_state_counts.csv",
    index=False
)

# Chart
plt.figure(figsize=(12,6))

state_counts.sort_values(
    "restaurant_count",
    ascending=False
).head(20).plot(
    x="province",
    y="restaurant_count",
    kind="bar"
)

plt.title("Top 20 States by Number of Fast Food Restaurants")
plt.tight_layout()

plt.savefig("charts/question3.png")

plt.show()

print("Question 3 completed")
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

os.makedirs("charts", exist_ok=True)

restaurants = pd.read_csv("fast_food_restaurants.csv")

# Count restaurants by city
city_counts = (
    restaurants.groupby("city")
    .size()
    .reset_index(name="restaurant_count")
)

# Major cities
major_cities = city_counts[
    city_counts["restaurant_count"] >= 5
]

top10 = major_cities.sort_values(
    "restaurant_count",
    ascending=False
).head(10)

bottom10 = major_cities.sort_values(
    "restaurant_count",
    ascending=True
).head(10)

print("TOP 10 CITIES")
print(top10)

print("\nBOTTOM 10 CITIES")
print(bottom10)

# Top chart
plt.figure(figsize=(10,5))

plt.bar(
    top10["city"],
    top10["restaurant_count"]
)

plt.xticks(rotation=45)
plt.title("Top 10 Cities by Fast Food Restaurants")

plt.tight_layout()

plt.savefig("charts/question5_top10.png")

plt.show()

# Bottom chart
plt.figure(figsize=(10,5))

plt.bar(
    bottom10["city"],
    bottom10["restaurant_count"]
)

plt.xticks(rotation=45)
plt.title("Bottom 10 Cities by Fast Food Restaurants")

plt.tight_layout()

plt.savefig("charts/question5_bottom10.png")

plt.show()

print("Question 5 completed")
```
