```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Create charts folder
os.makedirs("charts", exist_ok=True)

# Load datasets
restaurants = pd.read_csv("fast_food_restaurants.csv")
city_pop = pd.read_csv("city_population.csv")

# Check columns
print("Restaurant Columns:")
print(restaurants.columns)

print("\nCity Population Columns:")
print(city_pop.columns)

# Merge on city
merged = restaurants.merge(
    city_pop,
    on="city",
    how="inner"
)

# Save merged dataset
merged.to_csv(
    "merged_restaurant_data.csv",
    index=False
)

print("\nMerged Dataset Created")
print("Rows:", len(merged))

# Group by city and state
city_counts = (
    merged.groupby(["city", "state_name"])
    .agg(
        restaurant_count=("name", "count"),
        population=("population", "first")
    )
    .reset_index()
)

# Major cities only
major_cities = city_counts[
    city_counts["population"] >= 500000
].copy()

# Restaurants per 100k
major_cities["restaurants_per_100k"] = (
    major_cities["restaurant_count"]
    / major_cities["population"]
) * 100000

# Top 10
top_10 = (
    major_cities
    .sort_values(
        "restaurants_per_100k",
        ascending=False
    )
    .head(10)
)

# Bottom 10
bottom_10 = (
    major_cities
    .sort_values(
        "restaurants_per_100k",
        ascending=True
    )
    .head(10)
)

print("\nTOP 10 CITIES")
print(top_10)

print("\nBOTTOM 10 CITIES")
print(bottom_10)

# Save CSV results
top_10.to_csv(
    "charts/question5_top10.csv",
    index=False
)

bottom_10.to_csv(
    "charts/question5_bottom10.csv",
    index=False
)

# Top chart
plt.figure(figsize=(10,5))
plt.bar(
    top_10["city"],
    top_10["restaurants_per_100k"]
)

plt.title(
    "Top 10 Major Cities by Fast Food Restaurants Per 100,000 Residents"
)

plt.ylabel("Restaurants per 100k")
plt.xticks(rotation=45)
plt.tight_layout()

plt.savefig(
    "charts/question5_top10.png"
)

plt.show()

# Bottom chart
plt.figure(figsize=(10,5))
plt.bar(
    bottom_10["city"],
    bottom_10["restaurants_per_100k"]
)

plt.title(
    "Bottom 10 Major Cities by Fast Food Restaurants Per 100,000 Residents"
)

plt.ylabel("Restaurants per 100k")
plt.xticks(rotation=45)
plt.tight_layout()

plt.savefig(
    "charts/question5_bottom10.png"
)

plt.show()

print("\nFiles Created Successfully:")
print("charts/question5_top10.csv")
print("charts/question5_bottom10.csv")
print("charts/question5_top10.png")
print("charts/question5_bottom10.png")
```
