```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(
    "../data/processed/merged_restaurant_data.csv"
)

df.head()
```

```python
city_counts = (
    df.groupby("city")
      .size()
      .reset_index(name="restaurant_count")
)

city_counts.head()
```

```python
top_10 = (
    city_counts
    .sort_values(
        "restaurant_count",
        ascending=False
    )
    .head(10)
)

top_10
```

```python
city_counts["restaurant_count"].value_counts().head(20)
```

```python
bottom_10 = (
    city_counts[city_counts["restaurant_count"] > 1]
    .sort_values(
        "restaurant_count",
        ascending=True
    )
    .head(10)
)

bottom_10
```

```python
plt.figure(figsize=(10,5))

plt.bar(
    top_10["city"],
    top_10["restaurant_count"]
)

plt.title("Top 10 Cities by Number of Fast Food Restaurants")
plt.xlabel("City")
plt.ylabel("Restaurant Count")

plt.xticks(rotation=45)

plt.tight_layout()

plt.savefig(
    "../charts/question1_top_cities.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```
